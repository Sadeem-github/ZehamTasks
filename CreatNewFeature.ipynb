**Creating new Feature from Raw Data**

```python
import pandas as pd
```

```python
df=pd.read_csv('500_Person_Gender_Height_Weight_Index (1).csv')
```

```python
df.head()
```

I creat new feature (new coulmn) to calculate the **bmi**

bmi = weight รท Height^2 (in meter)

```python
# Convert hight from cm to m
# height / 100
df['Hight_M']=df['Height']/100
```

```python
df.head()
```

```python
df['BMI']=df['Weight']/(df['Hight_M']**2)
```
